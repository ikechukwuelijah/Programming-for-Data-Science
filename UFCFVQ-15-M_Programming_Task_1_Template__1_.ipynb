```python
def mean(numbers):
    return float(sum(numbers)) / max(len(numbers), 1)

mean([29, 17, 28, 6, 14, 7, 4, 27, 21, 15, 10, 16, 24, 26, 3, 11, 13, 8, 23, 9, 0, 22, 12, 2, 18, 19, 5, 1, 20, 25])
```

### Requirement FR2 - Develop a function to find the standard deviation

```python
def stdev(values):
    # get the mean
    mean_val = mean(values)

    # get the variance
    variance = sum([(x - mean_val) ** 2 for x in values]) / len(values)
    return variance ** 0.5

stdev([29, 17, 28, 6, 14, 7, 4, 27, 21, 15, 10, 16, 24, 26, 3, 11, 13, 8, 23, 9, 0, 22, 12, 2, 18, 19, 5, 1, 20, 25])
```

### Requirement FR3 - Develop a function to find the median 

```python
def median(values):

    # Sort List
    values = sorted(values)

    # Check for odd
    if len(values) % 2 == 1:
        return values[len(values) // 2]

    # Check for even
    else:
        lower = values[len(values) // 2 - 1]
        upper = values[len(values) // 2]
        return (float(lower + upper)) / 2

median([29, 17, 28, 6, 14, 7, 4, 27, 21, 15, 10, 16, 24, 26, 3, 11, 13, 8, 23, 9, 0, 22, 12, 2, 18, 19, 5, 1, 20, 25])
```

### Requirement FR4 - Develop a function to find the skewness

```python
def skewness(values):

    # Find the mean
    mean_val = mean(values)
    median_cal = median(values)
    standard_dev = stdev(values)
    skw = 3* (mean_val - median_cal) / standard_dev
    return  skw

skewness([29, 17, 28, 6, 14, 7, 4, 27, 21, 15, 10, 16, 24, 26, 3, 11, 13, 8, 23, 9, 0, 22, 12, 2, 18, 19, 5, 1, 20, 25])
```

### Requirement FR5 - Develop a function to read a single column from a CSV file

```python
def read_csv(file_name, column_number):
    # Read csv file
    with open(file_name, 'r') as f:

        # Get header columns
        header = f.readline()
        header = header.split(',')
        column_name = header[column_number]
        values = []
        for line in f:
            line = line.strip()
            line = line.split(',')
            values.append(line[column_number])

    return column_name, values
read_csv('task1.csv', 4)
```

### Requirement FR6 - Develop a function to read CSV data from a file into memory

```python
def read_csv_data_dict(file_name):
    # Read csv data
    with open(file_name) as csv_file:
        reader = csv_file.readlines()
        headers = reader[0].split(",")
        #Get the total column count
        total_columns = len(headers)
        my_dict = {}
    #Read each column to my_dict
    for col in range(0, total_columns):
        column_name, column_data = read_csv(file_name, col)
        my_dict[column_name] = column_data
    return my_dict

read_csv_data_dict('task1.csv')
```

### Requirement FR7 - Develop a function to generate a set of statistics for a given data file

```python
def csv_stats(file):
    csv_data = read_csv_data_dict(file)
    stats_vals = {}
    # Create the first dict values
    stats =  ["Mean","Stdev","Median","Skewness"]
    stats_vals.update({"Stats": stats})
    # get the statistics of the data
    for row, data in csv_data.items():
        val = [float(x) for x in data]
        arith_mean = round(mean(val),2)
        stdv = round(stdev(val),2)
        skew = round(skewness(val),2)
        median_val = round(median(val),2)
        # Build dict with stats values
        stats_vals.update({row: [arith_mean, stdv, median_val, skew]})
    return stats_vals
csv_stats('task1.csv')
```
